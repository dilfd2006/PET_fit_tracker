```python
#clean the data

df['Id'] = df['Id'].astype(str)
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'], format='%m/%d/%Y')
df.dtypes
```

```python
df['distance_diff'] = df['TotalDistance'] - df['TrackerDistance']
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.rename(columns={'activitydate':'activity_date', 'totalsteps':'total_steps', 'totaldistance':'total_distance', 'trackerdistance':'tracker_distance',
       'loggedactivitiesdistance':'logged_activities_distance', 'veryactivedistance':'very_active_distance',
       'moderatelyactivedistance':'moderately_active_distance', 'lightactivedistance':'light_active_distance',
       'sedentaryactivedistance':'sedentary_active_distance', 'veryactiveminutes':'very_active_minutes', 'fairlyactiveminutes':'fairly_active_minutes',
       'lightlyactiveminutes':'lightly_active_minutes', 'sedentaryminutes':'sedentary_minutes'}, inplace=True)
df.columns
```

```python
# create columns

day_of_week = df['activity_date'].dt.day_name()
df['day_of_week'] = day_of_week

df['n_day_of_week'] = df['activity_date'].dt.weekday + 1 #1 monday, 7 sunday
```

```python
# subset the data

df = df[['id', 'activity_date', 'total_steps', 'total_distance',
       'very_active_minutes', 'fairly_active_minutes',
       'lightly_active_minutes', 'sedentary_minutes', 'calories', 'day_of_week','n_day_of_week']].copy()
```

```python
df.head()
```

```python
# analysis

# steps categories
# sedentary: less than 5000 steps on average
# active: 5000 - 10000 steps on average
# very active: more than 10000 steps on average

df['id'].unique()
len(df['id'].unique())
```

```python
id_grp = df.groupby('id')
id_avg_step = id_grp['total_steps'].mean().sort_values(ascending=False)
id_avg_step = id_avg_step.to_frame()
```

```python
conditions = [
    (id_avg_step['total_steps'] <= 5000), (id_avg_step['total_steps'] > 5000) & (id_avg_step['total_steps'] < 10000), (id_avg_step['total_steps'] >= 10000)
]

values = ['sedentary', 'active', 'very_active']

id_avg_step['activity_level'] = np.select(conditions, values, default='')
id_activity_level = id_avg_step['activity_level']

df['activity_level'] = [id_activity_level[c] for c in df['id']]
df.head()
```

```python
# correlation between steps & calories burned

ax = sns.scatterplot(x='total_steps', y='calories', data=df, hue=df['activity_level'])

plt.title('Correlation calories vs. steps')
plt.tight_layout()
plt.show()
```

```python
# average steps per day
labels = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
fig, ax = plt.subplots(1, 1, figsize=(8,5))
day_grp = df.groupby('day_of_week')
avg_daily_steps = day_grp['total_steps'].mean()
avg_steps = df['total_steps'].mean()

plt.bar(avg_daily_steps.index, avg_daily_steps)
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels)

ax.axhline(y=avg_steps, color='blue', label='Avg daily steps')
ax.set_xlabel('Day of week')
ax.set_ylabel('Steps')
ax.set_title('Avg number of steps per day')
plt.legend()
plt.show()
```

```python
# percentage of activity in minutes

very_active_mins = df['very_active_minutes'].sum()
fairly_active_mins = df['fairly_active_minutes'].sum()
lighly_active_mins = df['lightly_active_minutes'].sum()
sedentary_mins = df['sedentary_minutes'].sum()

slices = [sedentary_mins, lighly_active_mins, fairly_active_mins, very_active_mins]

labels = ['sedentary_mins', 'lighly_active_mins', 'fairly_active_mins','very_active_mins']

explode = [0.1, 0, 0, 0]

plt.pie(slices, labels=labels, explode=explode, autopct='%1.1f%%')
plt.title('Percentage of activity level in minutes')
```

```python
# correlation between activity level in minutes & calories

labels = [1, 2, 3, 4, 5, 6, 7]

fig, axes = plt.subplots(2, 2, figsize=(8, 9), dpi=70)

sns.scatterplot(x='calories', y='sedentary_minutes', data=df, hue=df['activity_level'], ax=axes[0, 0], legend=False)
sns.scatterplot(x='calories', y='lightly_active_minutes', data=df, hue=df['activity_level'], ax=axes[0, 1], legend=False)
sns.scatterplot(x='calories', y='fairly_active_minutes', data=df, hue=df['activity_level'], ax=axes[1, 0], legend=False)
sns.scatterplot(x='calories', y='very_active_minutes', data=df, hue=df['activity_level'], ax=axes[1, 1], legend=True)

plt.legend(title='Activity level', fontsize=16, title_fontsize=18, bbox_to_anchor=(1.9, 1.8))
fig.suptitle('Correlation between activity level minutes and calories', fontsize=24)
plt.show()
```
